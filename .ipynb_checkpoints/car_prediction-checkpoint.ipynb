```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r'C:\Users\Raimu\PastaMain\Documentos\car_prediction_linear_sklearn\data.csv')
```

```python
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
df['Month'] = df['Year'].dt.month
df['Day'] = df['Year'].dt.day
df_selected = df[['Popularity', 'Year', 'Engine HP', 'Engine Cylinders', 'Month', 'Day', 'Market Category',
                   'Engine Fuel Type', 'Transmission Type', 'Driven_Wheels', 'Vehicle Size', 'Vehicle Style', 'MSRP']]
```

```python
# --- 1. Basic Data Overview ---
print("1. Data Shape (rows, columns):", df.shape)
print("\n2. First 5 rows:")
print(df.head())
print("\n3. Data Types:")
print(df.dtypes)
print("\n4. Missing Values:")
print(df.isnull().sum())
print("\n5. Descriptive Statistics:")
print(df.describe(include='all'))  
```

```python
# --- 2. Univariate Analysis ---

# 2.1 Target Variable (MSRP - Manufacturer's Suggested Retail Price)
plt.figure(figsize=(10, 6))
sns.histplot(df['MSRP'], kde=True, bins=50)
plt.title('Distribution of MSRP')
plt.xlabel('MSRP')
plt.ylabel('Frequency')
plt.show()
```

```python
numerical_cols = ['Popularity', 'Engine HP', 'Engine Cylinders', 'Month', 'Day']
imputer = SimpleImputer(strategy='mean')
df_selected.loc[:, numerical_cols] = imputer.fit_transform(df_selected[numerical_cols])
```

```python
categorical_cols = ['Market Category', 'Engine Fuel Type', 'Transmission Type', 
                    'Driven_Wheels', 'Vehicle Size', 'Vehicle Style']
imputer = SimpleImputer(strategy='most_frequent')
df_selected.loc[:, categorical_cols] = imputer.fit_transform(df_selected[categorical_cols])
```

```python
df_encoded = pd.get_dummies(df_selected, columns=['Market Category', 'Engine Fuel Type',
                                                   'Transmission Type', 'Driven_Wheels', 'Vehicle Size', 'Vehicle Style'], drop_first=True)
```

```python
df_train, df_test = train_test_split(df_encoded, test_size=0.2, random_state=42)
```

```python
missing_cols_test = set(df_train.columns) - set(df_test.columns)
for col in missing_cols_test:
    df_test[col] = 0
missing_cols_train = set(df_test.columns) - set(df_train.columns)
for col in missing_cols_train:
    df_train[col] = 0
df_test = df_test[df_train.columns]
```

```python
X_train = df_train.drop(['MSRP', 'Year'], axis=1)
y_train = df_train['MSRP']
X_test = df_test.drop(['MSRP', 'Year'], axis=1)
y_test = df_test['MSRP']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared: {r2:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")
```

```python
plt.figure(figsize=(6, 4))

sns.histplot(y_test, label='Actual', alpha=0.6, bins=100)
sns.histplot(y_pred, label='Predicted', alpha=0.8, bins=100)

plt.legend()

plt.ylabel('Frequency')
plt.xlabel('MSRP')
plt.title('Actual vs Predicted MSRP Distribution')

plt.show()
```

```python
# Create a new column `Engine HP Squared`
df_train['Engine HP Squared'] = df_train['Engine HP'] ** 2

# Print the first 5 rows of the updated training dataset
print(df_train.head().to_markdown(index=False, numalign="left", stralign="left"))

# Print the column names and their data types of the updated training dataset
print(df_train.info())
```

```python
# Create a new column `Engine HP Squared` in `df_test`
df_test['Engine HP Squared'] = df_test['Engine HP'] ** 2

# Separate features and target
X_train = df_train.drop(['MSRP', 'Year'], axis=1)
y_train = df_train['MSRP']
X_test = df_test.drop(['MSRP', 'Year'], axis=1)
y_test = df_test['MSRP']

# Create and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

# Print the evaluation metrics
print(f"Mean Squared Error: {mse:.2f}")
print(f"R-squared: {r2:.2f}")
print(f"Mean Absolute Error: {mae:.2f}")
```
